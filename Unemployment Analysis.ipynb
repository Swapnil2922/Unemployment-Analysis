# In order to examine the unemployment rate, let's now visualise the data. I'll start by looking at the approximate number of workers by India's various regions:

```python
user2.columns= ["Region","Date","Frequency",
               "Estimated Unemployment Rate(%)","Estimated Employed",
               "Estimated Labour Participation Rate(%)","Region.1",
               "longitude","latitude"]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed", hue="Region", data=user2)
plt.show()
```

```python
user1.columns= ["Region","Date","Frequency",
               "Estimated Unemployment Rate(%)","Estimated Employed",
               "Estimated Labour Participation Rate(%)","Area"]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed", hue="Region", data=user1)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate(%)", hue="Region", data=user2)
plt.show()
```
